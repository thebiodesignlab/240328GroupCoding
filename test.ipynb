```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# Path to the CSV file in the Documents folder
# Adjust 'your_file.csv' to the name of your CSV file
file_path = os.path.join(os.path.expanduser('~'), 'Documents', 'Codingschool', 'your_file.csv')

# Import the CSV file
data = pd.read_csv(file_path)

# Assuming the first column is for the X axis and the second column is for the Y axis
x = data.iloc[:, 1]
y = data.iloc[:, 0]

# Transform y to linearize the data: log(y) = log(a) + bx
y_transformed = np.log(y)

# Fit a line to the transformed data
coefficients = np.polyfit(x, y_transformed, 1)  # Linear fit to the transformed data
# Create a polynomial from the coefficients
polynomial = np.poly1d(coefficients)

# Generate x values spanning the min and max x values for the line
x_line = np.linspace(x.min(), x.max(), 100)
# Calculate y values for our line of best fit using the exponential of the polynomial
y_line = np.exp(polynomial(x_line))

plt.plot(x, y, marker='x', linestyle='none', label='Data Points')
# Plotting the line of best fit
plt.plot(x_line, y_line, color='red', label='Line of Best Fit')

# Plotting the data
plt.plot(x, y, marker='x')
plt.xlabel('Threshold Cycle')  # Replace 'X Axis Label' with your actual label
plt.ylabel('Phage Titre PFU/ml (plaque assay)')  # Replace 'Y Axis Label' with your actual label
plt.title('qPCR calibration curve')  # Replace with your actual title

# Setting the Y axis to logarithmic scale
plt.yscale('log')

plt.legend()

plt.savefig(os.path.join(os.path.expanduser('~'), 'Documents', 'Codingschool', 'graph.png'))

plt.show()

```
